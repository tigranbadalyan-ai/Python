
5. Сожмите строку (Run-Length Encoding): "aaabbc" → "a3b2c1".


```python
s = input()
result = ""
count = 1

for i in range(1, len(s)):
    if s[i] == s[i-1]:
        count += 1
    else:
        result += s[i-1] + str(count)
        count = 1

result += s[-1] + str(count)
print(result)
```


6. Восстановите строку из RLE-формата: "a3b2c1" → "aaabbc".

```python
s = input()
result = ""

for i in range(0, len(s), 2):
    result += s[i] * int(s[i+1])

print(result)
```

7. Выведите самое частое слово в тексте (без учёта регистра), при равенстве — самое раннее по появлению.


```python
text = input().lower().split()

freq = {}
for word in text:
    freq[word] = freq.get(word, 0) + 1

max_count = max(freq.values())

for word in text:
    if freq[word] == max_count:
        print(word)
        break
```

11. Запрашивайте числа, пока пользователь не введёт "stop". Выведите: min, max, среднее, количество.

```python
total = 0
count = 0
minimum = None
maximum = None

while True:
    x = eval(input("enter the number or 'stop': "))

    if x == "stop":
        break

    num = float(x)

    total += num
    count += 1

    if minimum is None or num < minimum:
        minimum = num

    if maximum is None or num > maximum:
        maximum = num

if count > 0:
    print("min:", minimum)
    print("max:", maximum)
    print("average:", total / count)
    print("count:", count)
else:
    print("there is not any number")
```

12. Реализуйте «угадай число» 1–100 с подсказками (Higher/Lower) и счётчиком попыток.

```python
import random

secret = random.randint(1, 100)
attempts = 0

while True:
    guess = int(input("enter the number from 1 to 100: "))
    attempts += 1

    if guess > secret:
        print("Lower")
    elif guess < secret:
        print("Higher")
    else:
        print("Correct!")
        print("Attempts:", attempts)
        break
```

14. Выведите все простые числа от 2 до n (эффективнее, чем проверка делителями до n).

```python
n_ = int(input('enter the number: '))
for num in range(2, n_ + 1):
    is_prime = True

    i = 2
    while i * i <= num:
        if num % i == 0:
            is_prime = False
            break
        i += 1

    if is_prime:    
        print(num)
```

15. Найдите НОД (GCD) двух чисел (алгоритм Евклида) и НОК (LCM).

```python
a = int(input("enter the first number: "))
b = int(input("enter the second number: "))

x = a
y = b


while y != 0:
    temp = y
    y = x % y
    x = temp

gcd = x

lcm = abs(a * b) // gcd

print("GCD:", gcd)
print("LCM:", lcm)
```

17. Найдите два наибольших и два наименьших элемента за один проход (без сортировки).

```python
my_ls = [1,2,3,4,5,6,7,6,5,4,4,3,2,8,9,0,11,22,33,44]
min1 = float('inf')
min2 = float('inf')
max1 = float('-inf')
max2 = float('-inf')
for x in my_ls:
    # minimums
    if x < min1:
        min2 = min1
        min1 = x
    elif x < min2:
        min2 = x
    
    # maximums
    if x > max1:
        max2 = max1
        max1 = x
    elif x > max2:
        max2 = x


print("Два наименьших:", min1, min2)
print("Два наибольших:", max1, max2)
```

22. Найдите все пары (a, b) из списка, для которых a + b = target (эффективно, не O(n^2)).

```python
numbers = input("Введите числа через пробел: ").split()

# превращаем в числа
i = 0
while i < len(numbers):
    numbers[i] = int(numbers[i])
    i += 1

target = int(input("Введите target: "))

seen = {}
result = []

i = 0
while i < len(numbers):

    num = numbers[i]
    need = target - num

    if need in seen:
        result.append((need, num))

    seen[num] = True

    i += 1

print("Пары:")
print(result)
```


24. Найдите главную и побочную диагональ (и их суммы).


```python
matrix = []
rows = int(input("Введите количество строк: "))
cols = int(input("Введите количество столбцов: "))

for _ in range(rows):
    row = list(map(int, input().split()))
    matrix.append(row)

arr = np.array(matrix)

main_diag = np.diagonal(arr)
print("Главная диагональ:", main_diag)
print("Сумма главной диагонали:", main_diag.sum())

anti_diag = np.diagonal(np.fliplr(arr))
print("Побочная диагональ:", anti_diag)
print("Сумма побочной диагонали:", anti_diag.sum())
```

27. Найдите координаты (row, col) максимального элемента матрицы.



```python
matrix = []
rows = int(input("Введите количество строк: "))
cols = int(input("Введите количество столбцов: "))

for _ in range(rows):
    row = list(map(int, input().split()))
    matrix.append(row)

arr = np.array(matrix)

max_val = np.max(arr)

# Индекс 
max_index = np.argmax(arr)

row, col = np.unravel_index(max_index, arr.shape)

print("Максимальный элемент:", max_val)
print("Координаты (row, col):", (row, col))
```

28. Постройте словарь частот слов в тексте (игнорируя регистр и пунктуацию), затем выведите топ-5 слов.
добавьте.

```python
import string

# Ввод текста
text = input("Введите текст: ")

# Приводим текст к нижнему регистру
text = text.lower()

# Убираем пунктуацию
for ch in string.punctuation:
    text = text.replace(ch, "")

# Разбиваем текст на слова
words = text.split()

# Считаем частоту слов
freq = {}
i = 0
while i < len(words):
    word = words[i]
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1
    i += 1


top_words = sorted(freq.items(), key=lambda x: x[1], reverse=True)[:5]

print("Топ-5 слов:")
for word, count in top_words:
    print(word, ":", count)
```


29. Инвертируйте словарь: value -> список ключей (если значения повторяются).


```python
d = {
    "a": 1,
    "b": 2,
    "c": 1,
    "d": 3,
    "e": 2
}

inv = {}

for key in d:
    value = d[key]
    if value in inv:
        inv[value].append(key)
    else:
        inv[value] = [key]

print("Инвертированный словарь:")
print(inv)
```

30. Смёрджите два словаря: если ключ общий, сложите значения (значения — числа), иначе просто добавьте.

```python
d1 = {"a": 2, "b": 3, "c": 5}
d2 = {"b": 4, "c": 1, "d": 7}

merged = {}


for key in d1:
    merged[key] = d1[key]

for key in d2:
    if key in merged:
        merged[key] += d2[key]  
    else:
        merged[key] = d2[key]   

print("Объединённый словарь:")
print(merged)
```
