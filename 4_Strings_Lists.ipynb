## String Format

As we learned in the Python Variables chapter, we cannot combine strings and numbers like this.

Как мы уже узнали в разделе о переменных Python, мы не можем напрямую объединять строки и числа таким способом.

```python
age = 36
txt = "My name is John, I am " + str(age)
print(txt)
```

But we can combine strings and numbers by using the format() method!
The format() method takes the passed arguments, formats them, and places them in the string where the placeholders {} are

Однако мы можем объединять строки и числа с помощью метода format()! Метод format() принимает переданные аргументы, форматирует их и подставляет в строку на места заполнителей {}.

```python
age = 36
txt = "My name is John, and I am {}"
print(txt.format(age))
```

The format() method takes unlimited number of arguments, and are placed into the respective placeholders.

Метод format() может принимать неограниченное количество аргументов, которые подставляются в соответствующие заполнители строки.

```python
quantity = 3
itemno = 567
price = 49.95
myorder = "I want {} pieces of item {} for {} dollars."
print(myorder.format(quantity, itemno, price))
```

You can use index numbers {0} to be sure the arguments are placed in the correct placeholders.

Вы можете использовать индексные номера {0}, {1}, {2} и т.д., чтобы гарантировать, что аргументы будут подставлены в нужные заполнители.

```python
quantity = 3
itemno = 567
price = 49.95
myorder = "I want to pay {2} dollars for {0} pieces of item {1}."
print(myorder.format(quantity, itemno, price))
```

We can do this with shorter code in this way:

Мы можем сделать то же самое более коротким и удобным способом следующим образом:

```python
quantity = 3
itemno = 567
price = 49.95
print(f"I want to pay {quantity} dollars for {itemno} pieces of item {price}.")
```

**Making copies of lists:** Making copies of lists is a little tricky due to the way Python handles lists.
Say we have a list `L` and we want to make a copy of the list and call it `M`. The expression `M = L` will
not work.

Копирование списков:
Создание копий списков может быть немного сложным из-за особенностей того, как Python работает со списками.
Предположим, у нас есть список L, и мы хотим создать его копию под именем M. Простое присваивание M = L не подойдёт.

```python
L = [1,2,3]
copy = L
L[0] = 9
print('L is now:', L, ' Copy:', copy)
```

We can see that the code did not work as we might have expected. When we changed `L`, the `copy` got changed as well. We can use the following code to actually copy (**shallow copy**) a list.

Мы видим, что код сработал не так, как мы ожидали. Когда мы изменили список L, копия также изменилась. Чтобы действительно создать копию списка (поверхностную копию, shallow copy), можно использовать следующий код.

```python
a = [7, 8]
L = [1,2,3]
# copy = L[:] # this line copies L
copy = L.copy()
L[0] = 9
print('L is now:', L, ' Copy:', copy)
```

```python
a = [7, 8]
L = [1,2,3, a]
copy = L.copy()
L[-1][0] = 9
print('L is now:', L, ' Copy:', copy)
```

```python
import copy
a = [7, 8]
L = [1,2,3, a]
copy = copy.deepcopy(L)
L[-1][0] = 0
L[0] = 9
print('L is now:', L, ' Copy:', copy)
print(id(L))
print(id(copy))
```

* Making a shallow copy of an object won't clone child objects. Therefore, the copy is not fully independent of the original.
* A deep copy of an object will recursively clone child objects. The clone is fully independent of the original, but creating a deep copy is slower.


* Поверхностная копия (shallow copy) объекта не копирует вложенные (дочерние) объекты. Поэтому такая копия не является полностью независимой от оригинала.

* Глубокая копия (deep copy) объекта рекурсивно копирует все вложенные объекты. Такая копия полностью независима от оригинала, однако её создание занимает больше времени.

The reason behind this behaviour is **referencing**.

Remember that everything in Python is an object. This includes **numbers**, **strings**, and **lists**. When we do a simple variable assignment, like `x=487`, what actually happens is Python creates an integer
object with the value `487`, and the variable `x` acts as a **reference to that object**. If we come along and declare `y=487`, then `y` also points to that same memory location.

When we set `L=[1,2,3]`, we create a list object `[1,2,3]` and a reference, `L`, to it. When we say
`copy=L`, we are making another reference to the object `[1,2,3]`. When we do `L[0]=9`, **because
lists are mutable**, the list `[1,2,3]` is changed in place to `[9,2,3]`. No new object is created.
The list `[1,2,3]` is now gone, and since copy is still pointing to the same location, it’s value is
`[9,2,3]`.

On the other hand, if we instead use `copy=L[:]`, we are actually **creating a new list object** somewhere
else in memory so that there are two copies of `[1,2,3]` in memory. Then when we do
`L[0]=9`, we are only changing the thing that `L` points to, and copy still points to `[1,2,3]`.

В Python все переменные являются ссылками на объекты в памяти. Поэтому при присваивании copy = L новый список не создаётся — обе переменные указывают на один и тот же объект, и изменения в L автоматически отражаются в copy. Чтобы создать независимую копию списка, нужно использовать L[:] или L.copy() (поверхностная копия — вложенные объекты остаются общими) либо copy.deepcopy() (глубокая копия — полностью независимая, но медленнее).

## Two-dimensional lists

We can nest lists and other object types within lists. For example, we can use nested lists to represent matrices

Мы можем вкладывать списки и другие типы объектов внутрь списков. Например, вложенные списки можно использовать для представления матриц.

```python
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

# or in one line
# matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
```

```python
[i[j] for i in matrix for j in range(len(i))]
```

With one index we get one row (a nested sublist)

Используя один индекс, мы получаем одну строку (вложенный подсписок).


```python
matrix[2]
```

With two indices we get an item within the row

Используя два индекса, мы можем получить элемент внутри строки.

```python
print(matrix[2][1])
print(matrix[1][1])
```

We can process the items of a two-dimensional lists using nested for loops

Мы можем обрабатывать элементы двумерных списков, используя вложенные циклы for.

```python
for row in matrix:
  for j in row:
    print(j, end=" ")
```
